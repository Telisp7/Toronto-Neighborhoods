<h1>Neighborhoods in Toronto</h1>

First, we import the libraries we will need

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import requests

```

We define the URL and send the GET request 

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
r = requests.get(url, headers=header)
```

Then, we transform the data into a dataframe

```python
tables = pd.read_html(r.text)

df=pd.DataFrame(tables[0])
```

The dataframe will have 3 columns

```python
df.columns=['Postcode','Borough','Neighbourhood']

df.drop([0],axis=0,inplace=True)

df.reset_index()
```

We drop the rows with unassigned values on the borough column

```python
df.drop(df[df['Borough']=="Not assigned"].index,axis=0, inplace=True)
```

Combine rows with different Neighborhood values but same Postal codes values.

```python
df1=df.groupby("Postcode").agg(lambda x:','.join(set(x)))
```

Assign the value of the borough column in the neighborhood column if neighborhood has "Not assigned" as a value

```python
df1.loc[df1['Neighbourhood']=="Not assigned",'Neighbourhood']=df1.loc[df1['Neighbourhood']=="Not assigned",'Borough']
```
